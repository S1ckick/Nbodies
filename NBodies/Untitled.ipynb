```python
import matplotlib.pyplot as plt
import numpy as np


import json
with open('log.json') as f:
  data_energy = json.load(f)

figure, axes = plt.subplots(figsize = (20,20))

axes.plot(data_energy["n"],data_energy["energy"],linestyle='solid', marker='o', markersize=4, label='Energy')

axes.set_xlabel('n computes')
axes.set_ylabel('energy')
axes.legend(prop={'size': 15})
```
