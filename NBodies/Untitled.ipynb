```python
import matplotlib.pyplot as plt
import numpy as np

figure, axes = plt.subplots(figsize = (20,20))

x, y = np.loadtxt('log.txt', delimiter=',', unpack=True)
axes.plot(x,y,linestyle='solid', marker='o', markersize=4, label='Theoretical solution')


axes.set_xlabel('n computes')
axes.set_ylabel('energy')
axes.legend(prop={'size': 15})
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
import numpy as np

x_1, y_1, z_1, x_2, y_2, z_2 = np.loadtxt('bodies.txt', delimiter=',', unpack=True)

axes3d.rcParams["legend.fontsize"] = 10
fig = plt.figure()
ax = fig.gca(projection="3d")


ax.plot(x_1, y_1, z_1, label="sun")
ax.plot(x_2, y_2, z_2, label="saturn")
ax.legend()
plt.show()
```
