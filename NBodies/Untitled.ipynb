```python
import matplotlib.pyplot as plt
import numpy as np

import json
with open('../log.json') as f:
  data_energy = json.load(f)
figure, axes = plt.subplots(figsize = (20,20))

axes.plot(list(map(float, data_energy)),linestyle='solid', marker='o', markersize=4, label='dE/E0')
axes.set_xlabel('n computes')
axes.set_ylabel('|E-E0|/E0')
axes.legend()

font = { 'size' : 40}

plt.rc('font', **font)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
import numpy as np

import json
with open('../bodies.json') as f:
    data_bodies = json.load(f)

coords = data_bodies["coords"]
iters = data_bodies["iter"]
bodies = data_bodies["num"]

fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')
#coords = np.reshape(data_bodies["coords"], (data_bodies["iter"], data_bodies["num"], 3))
#print(coords)
for i in range(bodies):
    x_body_coords = coords[i * 3 + 0 : : bodies * 3]
    y_body_coords = coords[i * 3 + 1 : : bodies * 3]
    z_body_coords = coords[i * 3 + 2 : : bodies * 3]
    ax.plot(list(map(float, x_body_coords)),list(map(float, y_body_coords)),list(map(float, z_body_coords)))

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.savefig('anime2.png', dpi=300)
```
